This notebook is for analyzing data from a survey about technology usage amonst college students.

## Import (and clean) data and packages

```python
import pandas as pd
import matplotlib.pyplot as plt

raw_tech_DF = pd.read_csv("student_tech_survey.csv")
cleaned_tech_DF = raw_tech_DF
cleaned_tech_DF["What\'s your major? (check all that apply)"] = cleaned_tech_DF["What\'s your major? (check all that apply)"].str.replace('"', '')  # Remove quotes
cleaned_tech_DF["What\'s your major? (check all that apply)"] = cleaned_tech_DF["What\'s your major? (check all that apply)"].str.replace(' and ', ',')  # Replace 'and' with a comma
cleaned_tech_DF["What\'s your major? (check all that apply)"] = cleaned_tech_DF["What\'s your major? (check all that apply)"].str.split(',')
cleaned_tech_DF["Which of the following devices do you own? (check all that apply)"] = cleaned_tech_DF["Which of the following devices do you own? (check all that apply)"].str.replace(" ","").str.split(",")
cleaned_tech_DF["Where do you prefer to study or work on academic assignments?"] = cleaned_tech_DF["Where do you prefer to study or work on academic assignments?"].str.split(',')
cleaned_tech_DF["How do you think technology has impacted your academic experience?"] = cleaned_tech_DF["How do you think technology has impacted your academic experience?"].str.split(',')

clean_len = len(cleaned_tech_DF)
```

## "Who?" Question Analysis

```python
def get_year_percent():
    year_counts = raw_tech_DF["What year are you?"].value_counts()
    year_percent = (year_counts / clean_len) * 100

    plt.figure(figsize=(4, 4))
    plt.pie(year_percent, labels=year_percent.index, autopct="%1.1f%%")
    plt.title('Distribution of College Years')
    plt.show()

get_year_percent()
```

```python
def get_major_percent():
    major_explode = cleaned_tech_DF.explode("What\'s your major? (check all that apply)")
    major_counts = major_explode["What\'s your major? (check all that apply)"].value_counts()
    major_percent = (major_counts / clean_len) * 100

    plt.figure(figsize=(4, 4))
    plt.pie(major_percent, labels=major_percent.index, autopct="%1.1f%%")
    plt.title('Distribution of College Majors')
    plt.show()

get_major_percent()
```

```python
def get_resident_percent():
    resident_counts = cleaned_tech_DF["Are you a:"].value_counts()
    resident_percent = (resident_counts / len(raw_tech_DF["Are you a:"])) * 100

    plt.figure(figsize=(4, 4))
    plt.pie(resident_percent, labels=resident_percent.index, autopct="%1.1f%%")
    plt.title('Distribution of College Residency Status')
    plt.show()

get_resident_percent()
```

## "What/Why" Question Analysis

```python
def get_device_percent():
    device_explode = cleaned_tech_DF.explode("Which of the following devices do you own? (check all that apply)")
    device_count = device_explode["Which of the following devices do you own? (check all that apply)"].value_counts()
    device_percent = (device_count / len(device_explode["Which of the following devices do you own? (check all that apply)"])) * 100

    plt.figure(figsize=(4, 4))
    plt.pie(device_percent, labels=device_percent.index, autopct="%1.1f%%")
    plt.title('Distribution of Device Ownership')
    plt.show()

get_device_percent()
```

```python
def get_primary_device_percent():
    primary_device_count = cleaned_tech_DF["What is your primary device for academic work?"].value_counts()
    resident_percent = primary_device_count / len(cleaned_tech_DF["What is your primary device for academic work?"]) * 100

    plt.figure(figsize=(4, 4))
    plt.pie(resident_percent, labels=resident_percent.index, autopct="%1.1f%%")
    plt.title('Distribution of Primary Device')
    plt.show()

get_primary_device_percent()
```

```python
def get_why_primary_device_percent():
    why_device_count = cleaned_tech_DF["Why do you use the primary device that you do?"].value_counts()
    why_device_percent = why_device_count / len(why_device_count) * 100

    plt.figure(figsize=(4, 4))
    plt.pie(why_device_percent, labels=why_device_percent.index, autopct="%1.1f%%")
    plt.title('Distribution of Why Primary Device Choice')
    plt.show()

get_why_primary_device_percent()
```

```python
def get_software_percent():
    software_count = cleaned_tech_DF["Which software applications do you use most frequently for academic purposes?"].value_counts()
    software_percent = software_count / len(cleaned_tech_DF["Which software applications do you use most frequently for academic purposes?"]) * 100

    plt.figure(figsize=(4, 4))
    plt.pie(software_percent, labels=software_percent.index, autopct="%1.1f%%")
    plt.title('Distribution of Software Usage')
    plt.show()

get_software_percent()
```

```python
def get_why_primary_software_percent():
    why_software_count = cleaned_tech_DF["Why is it your most frequently used software application?"].value_counts()
    why_software_percent = why_software_count / len(why_software_count) * 100

    plt.figure(figsize=(4, 4))
    plt.pie(why_software_percent, labels=why_software_percent.index, autopct="%1.1f%%")
    plt.title('Distribution of Why Software Choice')
    plt.show()

get_why_primary_software_percent()
```

```python
def get_hours_per_day():
    hours_count = cleaned_tech_DF["How many hours per day do you typically spend using electronic devices for academic or personal purposes?"].value_counts()
    hours_percent = hours_count / len(hours_count) * 100

    plt.figure(figsize=(4, 4))
    plt.pie(hours_percent, labels=hours_percent.index, autopct="%1.1f%%")
    plt.title('Distribution of Hours Spent on Devices')
    plt.show()

get_hours_per_day()
```
